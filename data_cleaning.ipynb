STEP 13: Create Charts 
 1. Bar Chart (Category Sales)

```python
import matplotlib.pyplot as plt

df.groupby('category')['total_sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

🔹 2. Top Products Chart

```python
df.groupby('product')['total_sales'].sum().sort_values().plot(kind='barh')
plt.title("Top Products")
plt.show()
```

3. Pie Chart

```python
df.groupby('category')['total_sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales Distribution")
plt.show()
```

### Business Insights
